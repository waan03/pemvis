```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat gambar bendera
flag_width = 1920
flag_height = 1080

# Membuat array kosong untuk bendera dengan ukuran yang sesuai
flag = np.ones((flag_height, flag_width, 3))

red_height = flag_height // 2

flag[:red_height, :, :] = [1, 0, 0]
flag[red_height:, :, :] = [1, 1, 1]


# Membuat plot
plt.imshow(flag)
plt.show()
```
